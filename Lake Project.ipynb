```python
df2 = df.loc[(df['Temp_Difference'] >= 1.0) & (df['depth'] != 0.0), ['lake', 'station_code', 'depth', 'date', 'Temp_Difference']]
df2
```

```python
df2.to_csv('Temperature greater than 1.csv')
```
